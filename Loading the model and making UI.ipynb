```python
from transformers import DistilBertTokenizer, TFDistilBertForSequenceClassification
import numpy as np

# Load the tokenizer and model
tokenizer = DistilBertTokenizer.from_pretrained('distilbert-base-uncased')
#loaded_model = TFDistilBertForSequenceClassification.from_pretrained('path/to/your/model/directory')

# Custom input text
custom_input = "Win free iphone by calling 6958659"

# Tokenize and preprocess the input
encoded_input = tokenizer(custom_input, truncation=True, padding=True, return_tensors='tf')

# Extract input tensors from batch encoding
input_ids = encoded_input['input_ids']
attention_mask = encoded_input['attention_mask']

# Make predictions
predictions = loaded_model(input_ids, attention_mask=attention_mask)

# Interpret predictions
predicted_label = np.argmax(predictions.logits.numpy())

# Map label to human-readable class


print(predicted_label)
```

```python
def classify(text):
  # Tokenize and preprocess the input
  encoded_input = tokenizer(text, truncation=True, padding=True, return_tensors='tf')

  # Extract input tensors from batch encoding
  input_ids = encoded_input['input_ids']
  attention_mask = encoded_input['attention_mask']

  # Make predictions
  predictions = loaded_model(input_ids, attention_mask=attention_mask)

  # Interpret predictions
  predicted_label = np.argmax(predictions.logits.numpy())

  # Map label to human-readable class


  if predicted_label==1:
    return "the message is spam"
  else:
    return "the message is not spam"
```

```python


import gradio as grd

# Note: Input samples can be an array of strings as there's only one input field

app_title = "Spam Classification Using DistilBert"

# Function to handle the text input and return the classification
def handle_text(input_text):
    return classify(input_text)

# Setting up the Gradio interface
app_interface = grd.Interface(
    handle_text,
    grd.Textbox(lines=3, label="Enter Text"),
    grd.Textbox(lines=3, label="Classification: Spam/Not Spam"),
    title=app_title,
    share=True
)

app_interface.launch()
```
